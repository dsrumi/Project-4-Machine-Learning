```python


def plot_distribution_by_category(df, col1, col2, ax):
    sns.countplot(data=df, y=col2, hue=col1, palette='coolwarm', 
                  order=df[col2].value_counts().index, ax=ax)
    
    # Adding labels to each bar
    for container in ax.containers:
        ax.bar_label(container, fmt='%d', label_type='edge', padding=3)
    
    # Setting titles and labels
    ax.set(title=f'{col1} Distribution by {col2}', xlabel='Count', ylabel=col2)
    ax.legend(title=col1)

# Define columns to plot
columns_to_plot = [('Attrition', 'Department'), ('Attrition', 'JobRole'),
                   ('Attrition', 'EducationField'),('Attrition', 'BusinessTravel'),
                   ('Attrition', 'Gender'),]

# Create a 2x2 subplot
fig, axes = plt.subplots(3, 2, figsize=(14, 10))  # 2 rows, 2 columns

# Iterate through the columns and plot
for ax, (col1, col2) in zip(axes.flatten(), columns_to_plot):
    plot_distribution_by_category(df, col1, col2, ax)

plt.tight_layout()  # Adjust spacing
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Select numerical columns
numerical_columns = df.select_dtypes(include=['number']).columns

# Calculate the correlation matrix
corr = df[numerical_columns].corr()

# Set up the figure size
plt.figure(figsize=(14, 10))

# Create a mask to hide the upper triangle (since correlation is symmetric)
mask = np.triu(np.ones_like(corr, dtype=bool))

# Draw the heatmap with better styling
sns.heatmap(corr, mask=mask, annot=True, cmap='coolwarm', fmt='.2f', 
            linewidths=0.5, annot_kws={"size": 10}, center=0, vmin=-1, vmax=1)

# Improve readability
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(fontsize=12)

# Add title
plt.title('Correlation Heatmap', fontsize=15, fontweight='bold')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Select numerical columns
numerical_columns = df.select_dtypes(include=['number']).columns

# Calculate the correlation matrix
corr = df[numerical_columns].corr()

# Plot the heatmap
plt.figure(figsize=(12,8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```


### **1. Understanding Correlation Values**
- The numbers in the heatmap represent **correlation coefficients** (ranging from -1 to +1).  
- **Color scale**:  
  - **Red (closer to +1)** → Strong positive correlation (as one variable increases, the other tends to increase).  
  - **Blue (closer to -1)** → Strong negative correlation (as one variable increases, the other tends to decrease).  
  - **White / Light colors (closer to 0)** → Weak or no correlation.  

### **2. Key Observations in Your Heatmap**
- **Diagonal values (1.00)** → These are self-correlations, meaning each variable is perfectly correlated with itself.  
- **High negative correlation (e.g., -0.68 for TotalWorkingYears and another variable)** → If a feature is strongly negatively correlated with another, an increase in one may suggest a decrease in the other.  
- **High positive correlation (e.g., 0.95 for MonthlyIncome and JobLevel)** → These variables are likely related; for example, as Job Level increases, Monthly Income also rises.  
- **Weak correlations (~0.00 to 0.10)** → These indicate little to no linear relationship between the variables.  

### **3. How to Use This for Analysis**
- **Feature Selection**: If two features have a very high correlation (e.g., JobLevel & MonthlyIncome), you might consider dropping one to avoid multicollinearity in predictive models.  
- **Understanding Attrition Drivers**: If you included `Attrition` in your heatmap, look at which variables have the strongest correlations (positive or negative).  
- **Data Insights**: Strong correlations (positive or negative) can reveal trends in HR data, such as the effect of tenure, salary, or job role on employee behavior.
