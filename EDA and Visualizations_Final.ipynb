```python


def plot_distribution_by_category(df, col1, col2, ax):
    sns.countplot(data=df, y=col2, hue=col1, palette='coolwarm', 
                  order=df[col2].value_counts().index, ax=ax)
    
    # Adding labels to each bar
    for container in ax.containers:
        ax.bar_label(container, fmt='%d', label_type='edge', padding=3)
    
    # Setting titles and labels
    ax.set(title=f'{col1} Distribution by {col2}', xlabel='Count', ylabel=col2)
    ax.legend(title=col1)

# Define columns to plot
columns_to_plot = [('Attrition', 'Department'), ('Attrition', 'JobRole'),
                   ('Attrition', 'EducationField'),('Attrition', 'BusinessTravel'),
                   ('Attrition', 'Gender'),('Attrition','MaritalStatus')]

# Create a 2x2 subplot
fig, axes = plt.subplots(3, 2, figsize=(14, 10))  # 2 rows, 2 columns

# Iterate through the columns and plot
for ax, (col1, col2) in zip(axes.flatten(), columns_to_plot):
    plot_distribution_by_category(df, col1, col2, ax)

plt.tight_layout()  # Adjust spacing
plt.show()
```

```python
# Select numerical columns
numerical_columns = df.select_dtypes(include=['number']).columns

# Calculate the correlation matrix
corr = df[numerical_columns].corr()

# Set up the figure size
plt.figure(figsize=(14, 10))

# Create a mask to hide the upper triangle (since correlation is symmetric)
mask = np.triu(np.ones_like(corr, dtype=bool))

# Draw the heatmap with better styling
sns.heatmap(corr, mask=mask, annot=True, cmap='coolwarm', fmt='.2f', 
            linewidths=0.5, annot_kws={"size": 10}, center=0, vmin=-1, vmax=1)

# Improve readability
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(fontsize=12)

# Add title
plt.title('Correlation Heatmap', fontsize=15, fontweight='bold')

# Show the plot
plt.show()
```


### **What This Graph Represents**  
This is a **correlation heatmap** that visualizes the relationships between different numerical  features in  IBM HR dataset. The values indicate how strongly two variables are correlated (positively or negatively).  

### **Key Observations**  
1. **Strong Positive Correlations (Red Cells)**  
   - Example: **YearsAtCompany vs. TotalWorkingYears** (~0.77)  
     -  Employees who have worked longer overall tend to have spent more years in their company.  
   - Example: **YearsWithCurrManager vs. YearsInCurrentRole** (~0.76)  
     - This suggests that employees tend to stay in their current roles for a similar duration as they stay with their current manager.  

2. **Strong Negative Correlations (Blue Cells)**  
   - Example: **YearsAtCompany vs. JobHopping Indicators**  
     - Employees with more years at the company tend to have fewer job changes, which aligns with expectations.  
   - Example: **JobRole Encodings vs. Other Features**  
     - Some job roles seem to have a negative correlation with specific factors like salary, department, or job satisfaction.  

3. **Weak or No Correlation (Near 0 Values)**  
   - Many of the smaller numbers (~0.00–0.10) indicate little to no linear relationship.  

### **How This Helps Your ML Model**  
- **Feature Selection**: If two independent variables are highly correlated, you may consider removing one to avoid multicollinearity.  
- **Predictor Insights**: Strong correlations (positive or negative) might indicate influential factors in attrition prediction.  
- **Scaling and Preprocessing**: Since some features have strong correlations, standardization or feature engineering could improve model accuracy.

# For Categorical Variables:

```python
# Create a contingency table for JobRole vs. Attrition
contingency_table = pd.crosstab(df["JobRole"], df["Attrition"])
print(contingency_table)
```

```python
from scipy.stats import chi2_contingency

# Run chi-square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

# Print results
print(f"Chi-square Statistic: {chi2}")
print(f"P-value: {p}")
```

```python
categorical_features = ["BusinessTravel", "Department", "EducationField", "Gender", "JobRole", "MaritalStatus"]

for feature in categorical_features:
    contingency_table = pd.crosstab(df[feature], df["Attrition"])
    chi2, p, dof, expected = chi2_contingency(contingency_table)
    print(f"Feature: {feature} | P-value: {p}")
```



| Feature         | P-value               |
|---------------|----------------------|
| BusinessTravel | 5.61e-06             |
| Department     | 0.00453              |
| EducationField | 0.00677              |
| Gender        | 0.29057              |
| JobRole       | 2.75e-15             |
| MaritalStatus | 9.46e-11             |



## Chi-Square Test Results Interpretation
Based onthe p-values, we can determine which categorical features have a statistically significant relationship with Attrition (assuming a significance threshold of 0.05).

### Key Findings:
 Significant Relationship with Attrition (p < 0.05)

#### BusinessTravel (p = 5.6e-06) → Employees' travel frequency impacts attrition.
#### Department (p = 0.0045) → Certain departments may have higher attrition.
#### EducationField (p = 0.0068) → Education background affects attrition.
#### JobRole (p = 2.75e-15) → The strongest relationship! Some job roles see much higher attrition.
#### MaritalStatus (p = 9.46e-11) → Marital status significantly influences attrition.
## No Significant Relationship (p ≥ 0.05)

Gender (p = 0.291) → Gender does not significantly impact attrition.


 ### Key Takeaways:
- Most significant features: JobRole, MaritalStatus, and BusinessTravel.
- Moderately significant features: Department and EducationField.
- Not significant: Gender (since p > 0.05), meaning gender likely does not impact attrition in this dataset.
