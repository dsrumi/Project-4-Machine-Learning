```python
import matplotlib.pyplot as plt
import seaborn as sns

def plot_distribution_by_category(df, col1, col2):
    plt.figure(figsize=(12, 4))
    
    # Bar Chart
    ax = sns.countplot(data=df, y=col2, hue=col1, palette='coolwarm',
                       order=df[col2].value_counts().index)
    
    # Adding labels to each bar
    for container in ax.containers:
        ax.bar_label(container, fmt='%d', label_type='edge', padding=3)
    
    # Setting titles and labels
    ax.set(title=f'{col1} Distribution by {col2}', xlabel='Count', ylabel=col2)
    plt.legend(title=col1)
    plt.tight_layout()
    plt.show()
```

```python
df = df.drop(['EmployeeNumber', 'EmployeeCount', 'Over18', 'StandardHours'], axis=1)
```

```python
if df is None:
    print("DataFrame is not loaded. Please check your data source.")
else:
    attrition_cat = df.select_dtypes(include='object').columns.tolist()
    unique_counts = df[attrition_cat].nunique()
    print(unique_counts)
```

```python
# Separate the target column (attrition) from the categorical columns
target = df['Attrition']
attrition_cat = [col for col in df.select_dtypes(include=['object']).columns if col != 'Attrition']
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
# Create a OneHotEncoder instance
enc = OneHotEncoder(sparse_output=False)
```

```python
# Fit and transform the OneHotEncoder using the categorical variable list
encode_df = pd.DataFrame(enc.fit_transform(df[attrition_cat]))
```

```python
# Add the encoded variable names to the dataframe
encode_df.columns = enc.get_feature_names_out(attrition_cat)
encode_df.head()
```

```python
# Merge one-hot encoded features and drop the originals
df = df.merge(encode_df,left_index=True, right_index=True)
df = df.drop(attrition_cat, axis=1)
df.head()
```

```python
y = df["Attrition"]
X = df.drop(columns="Attrition")
```

```python
y = y.map({'Yes': 1, 'No': 0})
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Split the data first (before scaling)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, stratify=y)

# Create the StandardScaler instance
scaler = StandardScaler()

# Fit the scaler on the training data and transform it
X_train_scaled = scaler.fit_transform(X_train)

# Transform the test data using the fitted scaler (without fitting again)
X_test_scaled = scaler.transform(X_test)
```

```python

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn import model_selection
import warnings

# Initialize models
lr = LogisticRegression()
gini = DecisionTreeClassifier(random_state=7)
forest = RandomForestClassifier(random_state=7)

# Define KFold cross-validation
kfold = model_selection.KFold(n_splits=10,shuffle= True, random_state=7)

# List of models
models = [('Logistic Regression', lr), ('Decision Tree', gini), ('Random Forest', forest)]

# Suppress warnings
warnings.simplefilter('ignore')

# Loop through each model and calculate cross-validated accuracy
for model in models:
    results = model_selection.cross_val_score(model[1], X_train, y_train, cv=kfold, scoring='accuracy')
    print(f"Cross validated Accuracy of {model[0]}: {results.mean():.3f}")
```

## Analysis:

The results  indicate the performance of three models: **Random Forest**, **Logistic Regression**, and **Decision Tree**. Here's how we can interpret the values for **Training Accuracy** and **Test Accuracy**:

### 1. **Training Accuracy**:
This is the accuracy of the model on the training data (`X_train` and `y_train`). It tells us how well the model fits the training data.

- **Random Forest**: 100% training accuracy. This suggests that the Random Forest model is overfitting, as it perfectly predicts the training data. Overfitting means the model has learned the details and noise in the training data, which may not generalize well to unseen data.
- **Logistic Regression**: 84.12% training accuracy. This indicates that the Logistic Regression model performs well on the training data but isn't perfectly fitted.
- **Decision Tree**: 100% training accuracy. Like Random Forest, this suggests the Decision Tree model has overfitted the training data.

### 2. **Test Accuracy**:
This is the accuracy of the model on the test data (`X_test` and `y_test`). It tells  how well the model generalizes to new, unseen data.

- **Random Forest**: 87.23% test accuracy. Even though the training accuracy is 100%, the Random Forest model generalizes well to unseen data (test set). This suggests it might have good generalization, despite overfitting on training data.
- **Logistic Regression**: 84.24% test accuracy. The Logistic Regression model has a similar performance on both the training and test data, suggesting it generalizes well without overfitting.
- **Decision Tree**: 77.17% test accuracy. Despite perfect accuracy on the training data, the Decision Tree performs worse on the test data, which indicates overfitting. It has learned too much from the training data, making it less capable of generalizing to new data.

### Key Takeaways:
1. **Random Forest** performs best in terms of test accuracy (87.23%), but has a potential overfitting issue. The high training accuracy (100%) means it fits the training data very well, but the test accuracy is slightly lower.
2. **Logistic Regression** has balanced performance on both the training and test sets, suggesting it may be the best model in terms of generalization.
3. **Decision Tree** overfits the training data, as shown by the perfect training accuracy (100%) but lower test accuracy (77.17%).

### Conclusion:
- **Random Forest** may be the best model in terms of test accuracy, but it’s important to check for overfitting (for example, by tuning hyperparameters or using cross-validation).
- **Logistic Regression** seems to generalize well and would be a safer choice if you're looking for a simpler, more interpretable model.
- **Decision Tree** should probably be pruned or adjusted to avoid overfitting.
